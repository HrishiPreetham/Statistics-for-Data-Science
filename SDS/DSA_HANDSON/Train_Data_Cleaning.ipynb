```python
#summary measures of a dataframe  column using statistics
import pandas as pd
import numpy as np
import statistics as st
df=pd.read_csv("train.csv")
```

```python
print('Mean',st.mean(df['ApplicantIncome'])) #prints the mean of a particular column note:This is only for quantifiable variables
```

```python
print('Median',st.median(df['ApplicantIncome'])) #prints the median of a particular column
```

```python
print('Standard deviation',st.pstdev(df['ApplicantIncome'])) #prints the population standard deviation
```

```python
print('Standard deviation',st.stdev(df['ApplicantIncome'])) #prints the sample standard deviation
```
