```python
import pandas as pd
from random import sample
df=pd.read_csv("train.csv")
```

```python
#simple random sampling
num=12
random_index=sample(range(df.shape[0]),num)
print(random_index)
print(df.iloc[random_index])
```

```python
#systematic sampling
k=20
ind=[i for i in range(df.shape[0]) if i%k==0]
print(ind)
print(df.iloc[ind])
```
