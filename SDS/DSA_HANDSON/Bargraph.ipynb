```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,2,3,4]
y=[2,4,5,3]
labels=['cars','bikes','bicycles','pedestrians']
#z=np.arange(0.2,100)
plt.bar(x,y,align='center')
plt.xticks(x,labels) #optional
#plt.yticks(x,z) #optional
plt.show()
```

```python
#horizontal bar graph
plt.barh(x,y,align='center')
plt.xticks(x,) #optional
plt.yticks(x,labels) #optional
plt.show()
```

```python
#edgealigned bar graph
plt.bar(x,y,align='edge')
plt.xticks(x,labels)
plt.show()
```

```python
plt.bar(x,y,align='edge',color='yellow')
plt.xticks(x,labels)
plt.show()
```

```python
#stack bar graphs
x=[1,2,3,4]
men=[4,7,8,5]
women=[2,5,9,10]
labels=['athletics','aquatics','hockey','tennis']
p1=plt.bar(x,men)
p2=plt.bar(x,women,bottom=men)
plt.xticks(x,labels)
plt.legend((p1[0],p2[0]),('Men','Women'))
plt.show()
```

```python
#grouped bar chart
x=np.arange(5)
y1=[34,56,12,89,67]
y2=[12,56,78,45,90]
width=0.4
plt.bar(x-0.2,y1,width)
plt.bar(x-0.2,y2,width)
plt.legend()
plt.show()
```
