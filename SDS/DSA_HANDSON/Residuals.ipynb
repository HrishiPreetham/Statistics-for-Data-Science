```python
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
import statistics
df=pd.read_csv("Residuals.csv")
plt.scatter(df['Wind Speed'],df.Current,color='red')
b1,b0=np.polyfit(df['Wind Speed'],df.Current,1)
y_p=b0+b1*np.array(df['Wind Speed'])
plt.plot(df['Wind Speed'],y_p)
plt.show()
```

```python
residuals=np.array(df.Current-y_p)
plt.scatter(y_p,residuals,color='red')
plt.plot([1,4],[0,0])
plt.show()
```

```python
#taking ln transformation
plt.scatter(np.log(df['Wind Speed']),df.Current,color='red')
b1,b0=np.polyfit(np.log(df['Wind Speed']),df.Current,1)
y_p=b0+b1*np.array(np.log(df['Wind Speed']))
plt.plot(np.log(df['Wind Speed']),y_p)
plt.show()
residuals=np.array(df.Current-y_p)
plt.scatter(y_p,residuals,color='red')
plt.plot([1,4],[0,0])
plt.show()
```

```python
plt.scatter(np.sqrt(df['Wind Speed']),df.Current,color='red')
b1,b0=np.polyfit(np.sqrt(df['Wind Speed']),df.Current,1)
y_p=b0+b1*np.array(np.sqrt(df['Wind Speed']))
plt.plot(np.sqrt(df['Wind Speed']),y_p)
plt.show()
residuals=np.array(df.Current-y_p)
plt.scatter(y_p,residuals,color='red')
plt.plot([1,4],[0,0])
plt.show()
```
